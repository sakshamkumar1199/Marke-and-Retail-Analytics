* we have taken median of differences between order_approved_at and order_purchased and then filled the na values with sum of order_purchased and diff
* we have taken median of differences between order_delivered_carrier_date and order_approved_at and then filled the nan values with sum of order_approved_at and diff

```python
# For order_approved_at

diff = (order['order_approved_at'] - order['order_purchase_timestamp']).median()
order['order_approved_at'].fillna(order['order_purchase_timestamp'] + diff,inplace=True)
```

```python
# For order_delivered_timestamp

diff = (order['order_delivered_timestamp'] - order['order_approved_at']).median()
order['order_delivered_timestamp'].fillna(order['order_approved_at'] + diff,inplace=True)
```

```python
order.isnull().sum()
```

```python
# creating one more column delayed or not where 0=within time and 1=delayed
order['delayed'] = np.where(order['order_delivered_timestamp']>order['order_estimated_delivery_date'],1,0)
```

```python
# creating new column ie. delivery time duration = time duration between delivered to customer date and purchase timestamp
order['delivery_time_duration'] = order['order_delivered_timestamp'] - order['order_purchase_timestamp']
```

### Merging table

```python
orded_prod =pd.merge(order_items, products, on="product_id", how="outer", indicator=True)
orded_prod.head()
```

```python
customer_ord =pd.merge(order, customers, on="customer_id", how="outer", indicator=True)
customer_ord.head()
```

```python
ord_payment = pd.merge(customer_ord, payments, on="order_id", how="outer")
ord_payment.head()
```

```python
final_order = pd.merge(ord_payment, orded_prod, on="order_id", how="outer")
final_order.head()
```

```python
final_order=final_order.drop({'_merge_x','_merge_y'},axis='columns')
```

```python
Market = pd.ExcelWriter('updated_retail_dataset.xlsx')

order.to_excel(Market, sheet_name = 'order', encoding= 'utf-8', index = False)
order_items.to_excel(Market, sheet_name = 'order_items', encoding= 'utf-8', index = False)
customers.to_excel(Market, sheet_name = 'customers', encoding= 'utf-8', index = False)
payments.to_excel(Market, sheet_name = 'payments', encoding= 'utf-8', index = False)
products.to_excel(Market, sheet_name = 'products', encoding= 'utf-8', index = False)
final_order.to_excel(Market, sheet_name = 'final_order', encoding= 'utf-8', index = False)

Market.save()
```
